Takes database of various cuisines and ingredients and outputs a txt file containing unique valid food items as ingredients.


```python
import pandas as pd
import json
```

```python
#the initial database was downloaded here: https://www.kaggle.com/datasets/kaggle/recipe-ingredients-dataset/data
#since the data is formated in categories of cuisine, the goal is to parse only the ingredients and create a list of unique valid ingredients

file = open("./train.json", 'r', encoding='utf8')
dataset_dict = json.load(file)
dataset = pd.DataFrame.from_dict(dataset_dict, orient='columns')
dataset.head(10)
```

```python
len(dataset)
```

```python
dataset.iloc[1]["ingredients"]
```

```python
#iterating through the database and adding all the ingredients to a set
ingredients = set()
for i in range(len(dataset)):
    row = dataset.iloc[i]["ingredients"]
    for j in row:
        ingredients.add(j)
```

```python
ingredients
```

```python
len(ingredients)
```

```python
"salt" in ingredients
```

```python
ingredients = list(ingredients)
```

```python
#writing the unique list of ingredients to an output file
with open('validIngredients.txt', 'w') as outfile:
    outfile.writelines(f"{str(i)}, " for i in ingredients[:-1])
    outfile.writelines(f"{str(ingredients[-1])}") #ensures that the last element doesn't also have a ', ' following it
```
